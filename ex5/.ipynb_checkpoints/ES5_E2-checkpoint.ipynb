```python
import numpy as np
import networkx as nx
from random import random, shuffle, seed
import matplotlib.pyplot as plt
```

# QUESTION 1
Define a function "generate_p1_p3_conf_model" which generates a configuration model of size $N$, where nodes have probability $p_1$ of having degree equal to 1, and probability $1-p_1$ of having degree 3.

# Answer 1

```python

seed(42)
def generate_p1_p3_conf_model(N, p1):
    # create an empty multigraph
    G = nx.empty_graph(N,create_using=nx.MultiGraph)
   
    # generate the degrees and check that the total sum of degrees is even. If this is not repeat the procedure of generating degrees
    # (We check that the sum of degrees is even to guarantee the respect of the handshaking lemma).
    even = False
    sum_k = 0
    while(not even):
        p3 = 1 - p1
        degs = (np.random.rand(N) > p1) * 2 + 1
        sum_k = np.sum(degs)
        even = (sum_k % 2) == 0
     
    # for the configuration model we draw pair of stubs (half-edges) uniformly at random
    # let's generate the array of stubs: in this array, each node i appears k_i times, where k_i is the degree of i.
    stubs = np.repeat(np.arange(N), degs)
   
    # to randomly match pairs of stubs, we can randomly shuffle the stub array and match element i and i+1 together for i in range(0,len(stub),2)
    np.random.shuffle(stubs)
    for i in range(0,len(stubs),2):
        G.add_edge(stubs[i], stubs[i+1])
        
    return G
    
# test
G = generate_p1_p3_conf_model(20, 0.3)
nx.draw(G)
```

# QUESTION 2
Generate an instance of such model for combinations of $N = 100, 1000, 5000, 10000$ and $p_1 = 0.25, 0.5, 0.75$. Compute in each case the proportion of self-edges and multiedges and the expected proportion of self-edges and multiedges based on the degree distribution.

# Answer 2

```python
# G.number_of_edges(v,u) gives number of edges between two nodes

for N in [100, 1000, 5000, 10000]:
    for p in [0.25, 0.5, 0.75]:
        p3 = (1-p)
        G = generate_p1_p3_conf_model(N, p)
        A = nx.to_numpy_array(G)
        loops = np.sum(A.diagonal())
        me = np.sum(nx.to_numpy_array(G)>=2) / 2
        
        # expected self loops, -> see notes for derivation
        loops_e = (p + 9*p3) / (p + 3*p3)
        me_e = (9*p3) / ((p + p3) ** 2 * N)
        print(loops, loops_e, me, me_e)
```

# QUESTION 3
Calculate the size of the largest component for a network of $N = 10000$ nodes with values of $p_1$ ranging from 0 to 1 in steps of 0.01. Then make a graph of the results as a function of $p_1$.

```python
f_x = np.zeros(100)
f_y = np.zeros(100)

for i in range(100):
    p = i * 0.01
    f_x[i] = p
    G = generate_p1_p3_conf_model(1000, p)
    cc= [len(c) for c in sorted(nx.connected_components(G), key=len, reverse=True)]
    f_y[i] = cc[0]
    
plt.plot(f_x, f_y)
```
